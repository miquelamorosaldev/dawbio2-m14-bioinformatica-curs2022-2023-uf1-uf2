# Màscares. 

## Màscares de Selecció (Boolean Indexing)

Referència:

<a href='https://pandas.pydata.org/docs/user_guide/10min.html#boolean-indexing'> 10min Pandas - Boolean Indexing</a>


### Mask 1. Seleccionar Pokemons criteri numèric. HP>= 100 

```python
mask_hp_over90 = pokedex.loc[ : , 'HP' ] >= 90

# Apliquem la màscara.
pokedex.loc[mask_hp_over90, : ]

#Comptem els que compleixen la màscara.
# pokedex.loc[mask_hp_over90,'HP'].count()
```


### Mask 2. Seleccionar Pokemons criteri booleà. Legendaris (Legendary = True)

```python
print(pokedex.loc[1,'Legendary'])

mask_legendaries = pokedex.loc[ : , 'Legendary' ] == True

# Apliquem la màscara.
pokedex.loc[mask_legendaries, : ]

#Comptem els que compleixen la màscara.
```

### Mask 3. Màscara d'Strings, composta.

Màscara d'Strings, composta.

**Ref:**

https://kanoki.org/2020/01/21/pandas-dataframe-filter-with-multiple-conditions/

#### Volem obtenir els pokemons que tinguin o bé el tipus 1 o el tipus 2 elèctric.


```python

mask_electrics = (pokedex.loc[ : , 'Type 1' ] == 'Electric') | (pokedex.loc[ : , 'Type 2' ] == 'Electric')

# Apliquem la màscara.
pokedex.loc[mask_electrics, : ]
```

## Màscares de substitució de valors.

Abans d'aplicar aquestes màscares fem una còpia del dataframe.

### Màscara Subs 1. Reemplaçar el tipus 1 i tipus 2 elèctric per un símbol amb un llamp 
🗲 => U+1F5F2.


```python
# Important fer una còpia de seguretat de tot el df abans de continuar.
import copy
pokedex2 = copy.deepcopy(pokedex)

# Màscara de substitució de valors.

# Pas 1. Aplicar la màscara per seleccionar els que són de tipus elèctric.
mask_electrics = (pokedex2.loc[ : , 'Type 1' ] == 'Electric') | (pokedex2.loc[ : , 'Type 2' ] == 'Electric')


# Pas 2. Aplicar la màscara; és a dir, modificar només les files del dataframe que compleixen la màscara.
#  df. 🗲 => U+1F5F2
pokedex2.loc[mask_electrics,['Type 1'] ] = "🗲"
pokedex2.loc[mask_electrics,['Type 2'] ] = "🗲"

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2.loc[mask_electrics, : ])

# Comentari, també converteix els que no tenen Type 2. Ara ho arreglarem.
```
